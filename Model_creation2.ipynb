```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('dataset/Disease_processed_data.csv')
df.head()
```

```python
df.columns
```

```python
# df['Disease'] = df['Disease'].str.join(' ','_')
df['Disease'] = df['Disease'].str.strip()
```

```python
df.info()
```

```python
df['Disease'] = df['Disease'].str.replace(' ','_')
```

```python
df1 = df[['Disease', 'Symptom_0', 'Symptom_1', 'Symptom_2', 'Symptom_3',
       'Symptom_4']]
df1.head()
```

```python
def fun1(columns):
    for i in columns:
        df1[i] = df1[i].str.strip()

fun1(['Disease', 'Symptom_0', 'Symptom_1', 'Symptom_2', 'Symptom_3',
       'Symptom_4'])
```

```python
df1['Disease'] = df1['Disease'].str.replace(' ','_')
```

```python
X = df1[['Symptom_0', 'Symptom_1', 'Symptom_2', 'Symptom_3',
       'Symptom_4']]
y = df1['Disease']
```

```python
X = pd.get_dummies(X,columns=['Symptom_0', 'Symptom_1', 'Symptom_2', 'Symptom_3',
       'Symptom_4'])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
def fun1(column):
    for i in column:
        X_train[i] = X_train[i].astype('int')

fun1(X_train.columns)
```

```python
def fun2(column):
    for i in column:
        X_test[i] = X_test[i].astype('int')

fun2(X_test.columns)
```

```python
X_train
```

```python
X_test
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr_classifier = LogisticRegression()
lr_classifier.fit(X_train,y_train)
```

```python
y_pred = lr_classifier.predict(X_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_pred,y_test)
```

Naive Baiyes

```python
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(X_train,y_train)
```

```python
y_pred = nb_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
metrics.accuracy_score(y_pred,y_test)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train,y_train)
```

```python
y_pred = dt_classifier.predict(X_test)
```

```python
metrics.accuracy_score(y_test,y_pred)
```

from these models we see that the naive bayes algorithm giving us the 87.34 % accuracy but the decision tree algorithm model gives us the 69.62 % accuracy , so we choose naive baise algorithm instead of using the decision tree algorithm .
